```python
# Q0_DICES.py
# StuID:B11030202   (revised: MEG304 2021Fall)
## TODO:  change this .py to a runnable jupyter notebook.
## GOAL:  Running 10-time(max.) games to get the sum of two-dices.
##        if the sum of the nth-trial is less than 6, then game over (Exit).
## 運行（最多）10次遊戲以獲得兩個骰子的總和。如果第n次試驗的總和小於6，則遊戲結束（退出）。

import random

def dice():
    dice1 = random.randint(1, 6)
    dice2 = random.randint(1, 6)
    return dice1, dice2, dice1 + dice2
```

```python
def Q1Task(n):
    print("Now, Max. Trails={}".format(n))
    for i in range(n):
        # 獲取骰子1、骰子2集骰子的和
        dice1, dice2, sum = dice()
        # 展示結果
        print("\tthe {}'th roll dices value {} = {} + {}".format(i+1, sum, dice1, dice2))
        # 若該次和小於6，中斷
        if sum < 6: break
```

```python
## --- your code /  brief  text-block (MD-Style) is here...
import random

def getDistance(pos1, pos2):
    distance = abs( pos1['x'] - pos2['x'] ) + abs( pos1['y'] - pos2['y'] )
    return distance

deliverers_name = ["甲外送員", "乙外送員"]
deliverers = list()
def genDeliverers(deliverers_name):
    for name in deliverers_name:
        deliverers.append({
            "name": name,
            "pos": {
                "x": random.randint(-1000, 1000),
                "y": random.randint(-1000, 1000)
            }
        })

genDeliverers(deliverers_name)
print("外送員資訊：")
print(deliverers)

customers = list()
print("在以下輸入3位顧客座標訊息(格式：X Y)：")
for i in range(3):
    input_data = input("{} 訂單(X, Y)> ".format(i + 1))
    input_data_slice = input_data.split(" ")
    try:
        x = int(input_data_slice[0])
        y = int(input_data_slice[1])
        pos = {
            "x": x,
            "y": y
        }
        print("")
        customers.append( {
            "name": "顧客" + chr(66 + i),
            "pos": pos
        } )
    except:
        print("輸入錯誤")
        break

print("顧客資訊")
print(customers)

max_deliverer = deliverers[0]
max_distance = None
for deliverer in deliverers:
    customers_for_a = sorted(customers, key = lambda customer: getDistance(customer['pos'], deliverer['pos']))
    full_distance = getDistance(customers_for_a[0]['pos'], deliverer['pos']) + getDistance(customers_for_a[1]['pos'], deliverer['pos'])
    print("{} 送餐到 {} 及 {} 的總距離為 {}".format(deliverer['name'], customers_for_a[0]['name'], customers_for_a[1]['name'], full_distance))
    if max_distance == None or full_distance > max_distance:
        max_distance = full_distance
        max_deliverer = deliverer
print("送餐路徑最長的為 {}".format(max_deliverer['name']))
```
